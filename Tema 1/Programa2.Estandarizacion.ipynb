```python
from sklearn import preprocessing
import numpy as np
```

```python
# Carga de datos.
X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])
```

```python
# Limpieza de datos: estandarización.
standardizer = preprocessing.StandardScaler()

# Estandarizamos los datos de entrenamiento
X_train_std = standardizer.fit_transform(X_train)
print(X_train_std) #valores negativos y positivos; y la media 0
```

```python
X_train_2 = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])
```

```python
# Aplicamos los métodos "fit" y "transform" por separado, esto es si tenemos un dato de test
f = standardizer.fit(X_train_2)
X_train_std_2 = f.transform(X_train_2)
print(X_train_std_2)
# el resultado nos sale el mismo
```

```python
# Comprobar si las matrices estandarizadas son iguales
X_train_std_2 == X_train_std
```
