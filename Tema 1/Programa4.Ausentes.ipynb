```python
import numpy as np
from sklearn.impute import SimpleImputer #se le puede dar distintas estrategias 
```

```python
# Carga de datos
X_train = [[1,      4], 
           [np.nan, 3], 
           [7,      8]]
```

```python
# Limpieza de datos: imputación valores ausentes (modelo)
imp = SimpleImputer(strategy='mean') # Utilizamos la estrategia de la media

# Aplicar los métodos "fit" y "transform" para imputar los valores pérdidos de X_train
imputer = imp.fit(X_train) #poenmos xtrain de nuevo pq si luego queremos aplicar a otros dataset y estamosa aplicando el valor medio, tiene que ser de train. no puede ser
X_train_clean = imputer.transform(X_train)
print(X_train_clean) #como hay un valor nam, se sustituye ese valor por la media de su columna 
```

```python
# Imputación de valores ausentes sobre el conjunto de test
X_test = [[np.nan, 10], 
          [6,      np.nan], 
          [8,      2]]

# Imputar los datos del conjunto de test
X_test_clean = imputer.transform(X_test)
print(X_test_clean) #no hace la media de esta matriz, si no de la anterior
```

Cuando usaste el objeto imputer (que probablemente es un SimpleImputer), este ya había sido entrenado previamente con un conjunto de datos de entrenamiento (X_train) mediante el método .fit().

La explicación técnica
Al hacer imputer.fit(X_train), el objeto calculó y almacenó internamente las medias de cada columna de X_train. Estas medias se guardan en un atributo llamado statistics_.

Cuando ejecutas X_test_clean = imputer.transform(X_test):

El imputer busca los valores NaN en tu matriz de test.

En lugar de calcular la media de X_test (que es lo que tú esperabas), el imputer rellena esos huecos con los valores que aprendió de X_train.

Este cuaderno explica la **imputación de valores ausentes**, una técnica de limpieza de datos que consiste en rellenar los huecos vacíos (`NaN`) con valores estimados para no tener que descartar muestras completas.

Aquí tienes el desglose de lo que sucede en el código:

---

## 1. El Herramienta: `SimpleImputer`

El código utiliza `SimpleImputer` de la librería `scikit-learn`. Este objeto permite elegir diferentes estrategias para rellenar datos faltantes; en este caso, se utiliza **`strategy='mean'`** (la media aritmética).

## 2. Flujo de Trabajo (Train vs. Test)

El cuaderno refuerza un concepto fundamental en Machine Learning que ya viste en la normalización: **la separación entre entrenamiento y prueba**.

### Fase de Entrenamiento (`X_train`)

1. **Datos:** Tienes una matriz donde falta un valor en la primera columna: `[np.nan, 3]`.
2. **`fit(X_train)`:** El imputer calcula la media de cada columna de entrenamiento.
* Para la primera columna: .


3. **`transform(X_train)`:** Sustituye el `NaN` por ese **4.0**.

### Fase de Prueba (`X_test`)

Aquí es donde ocurre lo más importante:

* Al ejecutar `imputer.transform(X_test)`, el programa rellena los huecos de los datos nuevos usando las medias que aprendió de `X_train`.
* **¿Por qué?** Porque en un escenario real, no conocerías la media de los datos futuros. Usar la media de entrenamiento garantiza que el modelo sea consistente y evita el **Data Leakage** (fuga de datos).

---

## 3. Resumen Técnico

* **`.fit()`**: El imputer calcula y guarda las medias en el atributo `statistics_`.
* **`.transform()`**: Aplica esos valores guardados para rellenar los `NaN` en cualquier matriz que le pases.
* **Resultado en Test**: Aunque en `X_test` la primera columna tiene los valores `6` y `8` (cuya media es 7), el imputer usará el **4.0** (la media de `X_train`) para rellenar el hueco.
